```python
# the first task
import matplotlib.image as mpimg 
import matplotlib.pyplot as plt
import numpy as np

#the function responsible for converting an rgba image to rgb
def rgba2rgb( rgba, background=(255,255,255) ):
    #storing the shape of the array
    row, col, ch = rgba.shape                    

    #if rgb return
    if ch == 3:
        return rgba
    
    #if rgba convert it
    assert ch == 4

    #seperating the channels
    rgb = np.zeros( (row, col, 3), dtype='float32' )
    r, g, b, a = rgba[:,:,0], rgba[:,:,1], rgba[:,:,2], rgba[:,:,3]

    a = np.asarray( a, dtype='float32' ) / 255.0

    R, G, B = background

    #calculating the rgb channels
    rgb[:,:,0] = r * a + (1.0 - a) * R
    rgb[:,:,1] = g * a + (1.0 - a) * G
    rgb[:,:,2] = b * a + (1.0 - a) * B

    return np.asarray( rgb, dtype='uint8' )

imgs = []
#where the images are stored
location = 'dl_khf1_pictures/'
np.set_printoptions(threshold=5)

#opening images and converting it to rgb
for i in range(5):
    imgs.append(mpimg.imread(location + str(i + 1) + '.png'))
    imgs[i] = imgs[i] * 255
    imgs[i] = rgba2rgb(imgs[i])
    
#showing the images and prining out the rgb channels
for i in range(5):
    plt.figure()
    plt.imshow(imgs[i][:, :, :])
    print("\nthe number " + str(i + 1) + " images content:")
    print("r:")
    print(imgs[i][:, :, 0])
    print("g:")
    print(imgs[i][:, :, 1])
    print("b:")
    print(imgs[i][:, :, 2])
    
```

```python
# the second task

#seperating and flattening the image arrays into a 1 dimensional array
flattened_imgs = []
r = []
g = []
b = []
for i in range(5):
    r.append(np.array(imgs[i][:, :, 0].flatten()))
    g.append(np.array(imgs[i][:, :, 1].flatten()))
    b.append(np.array(imgs[i][:, :, 2].flatten()))

#calculating the mean values
print("the mean values:")
for i in range(5):
    print("\nnumber " + str(i + 1) + " picture:")
    print(np.mean(r[i]))
    print(np.mean(g[i]))
    print(np.mean(b[i]))

#calculating the standard deviation values
print("\n\nthe standard deviation values:")
for i in range(5):
    print("\nnumber " + str(i + 1) + " picture:")
    print(np.std(r[i]))
    print(np.std(g[i]))
    print(np.std(b[i]))

#substracting the mean values of the datasets so the expected value becomes 0
for i in range(5):
    r[i] = r[i] - np.mean(r[i])
    g[i] = g[i] - np.mean(g[i])
    b[i] = b[i] - np.mean(b[i])
    
#dividing by the standard deviation so it becomes 1
for i in range(5):
    r[i] = r[i] / np.std(r[i])
    g[i] = g[i] / np.std(g[i])
    b[i] = b[i] / np.std(b[i])

#printing out everything again
print("\n\nAfter the change:")
print("the mean values:")
for i in range(5):
    print("\nnumber " + str(i + 1) + " picture:")
    print(np.mean(r[i]))
    print(np.mean(g[i]))
    print(np.mean(b[i]))
    
print("\n\nthe standard deviation values:")
for i in range(5):
    print("\nnumber " + str(i + 1) + " picture:")
    print(np.std(r[i]))
    print(np.std(g[i]))
    print(np.std(b[i]))
```

```python
# the third task

from scipy.io import wavfile

#reading the wav file and extracting the data to samplerate and data
samplerate, data = wavfile.read('dl_khf1_sound/cantinaband.wav')
length = data.shape[0] / samplerate

#plotting the audio file as a spectogram
time = np.linspace(0, length, data.shape[0])
plt.plot(time, data, label = "Values")
plt.legend()
#setting the axis names
plt.xlabel("Time [s]")
plt.ylabel("Amplitude")
plt.show()
```

```python
# the fourth task

import bs4
import requests
import re

#getting the deep learning url
url = 'https://en.wikipedia.org/wiki/Deep_learning'

#getting the url response and parsing it with html parser
response = requests.get(url)
soup = bs4.BeautifulSoup(response.content, 'html.parser')

#getting the text only
text = soup.get_text()

#the dictionary that contains the frequency of the letters
letter_freq = {} 
 
#filling up the dictionary with the characters in the string
#just by adding +1 to the values of the letters
for i in str(text).lower(): 
    if i in letter_freq: 
        letter_freq[i] += 1
    else: 
        letter_freq[i] = 1

#making a list of characters that are not letters 
#(sadly the isalpha() function didnt work so had to use regex)
remove = []
for letter in letter_freq:
    if  re.match("^[a-z]", letter) is None:
        remove.append(letter)

#removing the characters that are not letters from the dictionary made by the list above     
for i in remove: letter_freq.pop(i)     
    
#finally plotting the dictionary using a histogram
plt.bar(list(letter_freq.keys()), letter_freq.values(), color='g')
plt.show()
```
